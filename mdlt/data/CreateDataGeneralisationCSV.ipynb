```python
import pandas as pd
import numpy as np
Dataset='VLCS'
filename=Dataset+'/'+Dataset+'.csv'
data=pd.read_csv(filename,header=[0])
data.reset_index()
data.drop('split',axis=1)
```

```python
def data4DG(data,Dataset):
######################################
# For every unique class in env, we create a different csv file 
#####################################
    
    Files=data['env'].unique()
    for i in Files:
        data1=data[:]
        data1['split']=np.where(data1.env==i,'val','train')
        start=data1[data1['env']==i].iloc[0].name
        end=data1[data1['env']==i].iloc[-1].name+1
        data1['split'][start:end]=np.where(data1['split'][start:end].index % 2 ==0,'val','test')        
        
        filename=Dataset+'/'+Dataset+'_'+i+'.csv'
        data1.to_csv(filename) 
        
    
```

```python
data4DG(data,Dataset)
```
